## Using NLTK build a ML classifier for movie reviews

```python
import pandas as pd
import numpy as np
import os
from sklearn.feature_extraction.text import CountVectorizer
```

```python
os.listdir(nltk.data.find('corpora'))
```

```python
nltk.download('movie_reviews')
from nltk.corpus import movie_reviews
```

```python
movie_reviews.categories()
```

```python
len(movie_reviews.fileids('pos'))
```

```python
pos_rev = movie_reviews.fileids('pos')
```

```python
len(movie_reviews.fileids('neg'))
```

```python
neg_rev = movie_reviews.fileids('neg')
```

```python
neg_rev
```

```python
len(movie_reviews.fileids())
```

```python
rev = movie_reviews.words('pos/cv000_29590.txt')
```

```python
neg_rev
```

```python
nltk.corpus.movie_reviews.words('neg/cv000_29416.txt')
```

```python
rev_list = []
rev_types = [neg_rev, pos_rev]
for type_ in rev_types:
    for rev in type_:
        words = nltk.corpus.movie_reviews.words(rev)
        string_words = ' '.join(words)
        string_words = string_words.replace(',', '')
        string_words = string_words.replace("\'", "")
        rev_list.append(string_words)

    
    
```

```python
len(rev_list)
```

```python
## because the first 1000 are negative reviews and the last 1000 are positive ones
neg_target = np.zeros((1000,), dtype=np.int)
pos_target = np.ones((1000,), dtype=np.int)
```

```python
target_list = []
for target in neg_target:
    target_list.append(target)
for target in pos_target:
    target_list.append(target)
```

```python
y = pd.Series(target_list)
```

```python
y.head()
```

```python
y.tail()
```

```python
y2 = [0] * 1000 + [1] * 1000
y2 = pd.Series(y2)
```

```python
y2.head()
```

```python
y2.tail()
```

```python
count_vect = CountVectorizer(lowercase=True, stop_words='english', min_df=2)
```

```python
X_count_vect = count_vect.fit_transform(rev_list)
```

```python
X_count_vect.shape
```

```python
X_names = count_vect.get_feature_names()
```

```python
X_names[:10]
```

```python
type(X_count_vect)
```

```python
X_count_vect = pd.DataFrame(X_count_vect.toarray(), columns = X_names)
```

```python
X_count_vect
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_count_vect,y, test_size=0.25, random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb_pred_y = gnb.fit(x_train, y_train).predict(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
clf = MultinomialNB()
clf_pred = clf.fit(x_train, y_train).predict(x_test)
```

```python
metrics.accuracy_score(y_test, clf_pred)
```

```python
metrics.accuracy_score(y_test, gnb_pred_y)
```

```python
confusion_matrix(y_test, gnb_pred_y)
```

```python
confusion_matrix(y_test, clf_pred)
```
